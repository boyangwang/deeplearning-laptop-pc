````python
myvgg = Vgg16()
````

````python
SAMPLE_PATH = './sample/'
batches = myvgg.get_batches(SAMPLE_PATH+'train', batch_size=16)
val_batches = myvgg.get_batches(SAMPLE_PATH+'valid', batch_size=16)
myvgg.finetune(batches)
myvgg.fit(batches, val_batches, nb_epoch=3)
#preds = vgg.test('./test/', batch_size = batch_size*2)
````

````python
myvgg.fit(batches, val_batches, nb_epoch=3)
````

````python
val_batches, probs = myvgg.test('./valid/', batch_size = 16)
filenames = val_batches.filenames
expected_labels = val_batches.classes #0 or 1
````

````python
probs[:, 0] # means keep only element index 0
our_predictions = 1 - probs[:,0]
our_labels = np.round(our_predictions)
````

````python
def to_plot(img):
    if K.image_dim_ordering() == 'tf':
        return np.rollaxis(img, 0, 1).astype(np.uint8)
    else:
        return np.rollaxis(img, 0, 3).astype(np.uint8)
def plot(img):
    plt.imshow(to_plot(img))
def plots(ims, figsize=(12,6), rows=1, interp=False, titles=None):
    if type(ims[0]) is np.ndarray:
        ims = np.array(ims).astype(np.uint8)
        if (ims.shape[-1] != 3):
            ims = ims.transpose((0,2,3,1))
    f = plt.figure(figsize=figsize)
    for i in range(len(ims)):
        sp = f.add_subplot(rows, len(ims)//rows, i+1)
        sp.axis('Off')
        if titles is not None:
            sp.set_title(titles[i], fontsize=16)
        plt.imshow(ims[i], interpolation=None if interp else 'none')
````

````python
def plots_idx(idx, titles=None):
    plots([image.load_img('./valid/' + filenames[i]) for i in idx], titles=titles)
````

````python
incorrect_cats = np.where((our_labels==0) & (our_labels!=expected_labels))[0]
print("Found %d incorrect, we think laptop but is PC" % len(incorrect_cats))
if len(incorrect_cats):
    most_incorrect_cats = np.argsort(our_predictions[incorrect_cats])[::-1][:8]
    plots_idx(incorrect_cats[most_incorrect_cats], our_predictions[incorrect_cats][most_incorrect_cats])
````
